```python
X = df.iloc[:, :-1].values
Y = df.iloc[:, -1].values
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, shuffle = False)
```

```python
print(X_train)
```

```python
print(len(X_train),len(X_test))
```

```python
lr = LinearRegression().fit(X_train,Y_train)
```

```python
Y_pred = lr.predict(X_test)
```

```python
#  RMSE 
rmse = 0

for i in range(len(Y_pred)):
  rmse += pow((Y_pred[i]-Y_test[i]),2)
rmse = rmse / len(Y_pred)
rmse = pow(rmse,0.5)


print("RMSE for Linear regression:",rmse)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8,5]
plt.plot(Y_pred,label = "Predicted Closing Price")
# plt.show()
plt.title("Closing price for Stock vs Datapoints for Linear Regression")
plt.xlabel("Datapoints")
plt.ylabel("Closing price for Stock")
plt.plot(Y_test,label = "Actual Closing Price")
plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
dtr = DecisionTreeRegressor(random_state=0)
# cross_val_score(dtr, X, Y, cv=10)

Y_pred = dtr.fit(X_train, Y_train).predict(X_test)
```

```python
print(Y_pred)
```

```python
#  RMSE 
rmse = 0

for i in range(len(Y_pred)):
  rmse += (Y_pred[i]-Y_test[i])*(Y_pred[i]-Y_test[i])

rmse = rmse / len(Y_pred)
rmse = pow(rmse,0.5)

print("RMSE for Decision Tree:",rmse)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8,5]
plt.plot(Y_pred,label = "Predicted Closing Price")
# plt.show()
plt.title("Closing price for Stock vs Datapoints for Decision Tree")
plt.xlabel("Datapoints")
plt.ylabel("Closing price for Stock")
plt.plot(Y_test,label = "Actual Closing Price")
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=0)
Y_pred = rfr.fit(X_train, Y_train).predict(X_test)
```

```python
#  RMSE 
rmse = 0

for i in range(len(Y_pred)):
  rmse += (Y_pred[i]-Y_test[i])*(Y_pred[i]-Y_test[i])

rmse = rmse / len(Y_pred)
rmse = pow(rmse,0.5)

print("RMSE for Random Forest:",rmse)
```
