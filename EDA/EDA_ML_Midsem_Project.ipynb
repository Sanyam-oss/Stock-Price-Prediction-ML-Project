```python
vol = df['Volume'].to_list()

vol = np.array(vol)
# vol = (vol/(5*1e6))+30

fig, ax = plt.subplots()

ax.grid()

ax.plot(df['Close'].to_list(), marker="o", markersize=1, color="black", linewidth=1, markeredgecolor="black", label = 'Close Price')
ax.tick_params(axis='y', labelcolor='black')

ax2 = ax.twinx()
ax2.plot(vol, marker="o", markersize=1, color="blue", linewidth=1, markeredgecolor="blue", label = 'Volume')
ax2.tick_params(axis='y', labelcolor='Blue')

ax2.legend()

ax.set_title("Volume v/s Stock Price")
```

```python
sentiment_score = []
total_tweets = []
change = []

for i in range(len(df['Positive5'].to_list())) :

  sentiment_score.append(df['Positive5'].to_list()[i] - df['Negative5'].to_list()[i] + df['Positive4'].to_list()[i] - df['Negative4'].to_list()[i] + df['Positive3'].to_list()[i] - df['Negative3'].to_list()[i] )
  total_tweets.append((df['Positive5'].to_list()[i] + df['Negative5'].to_list()[i] + df['Neutral5'].to_list()[i] + df['Positive4'].to_list()[i] + df['Negative4'].to_list()[i] + df['Neutral4'].to_list()[i] + df['Positive3'].to_list()[i] + df['Negative3'].to_list()[i] + df['Neutral3'].to_list()[i])/3 )
  change.append((df['Close'].to_list()[i] - df['Close5'].to_list()[i]))

```

```python
plt.rcParams["figure.figsize"] = [10.00, 5.00]
plt.rcParams["figure.autolayout"] = True

# plt.xlim(xlim[0], xlim[1])
# plt.ylim(ylim[0], ylim[1])
plt.grid()

plt.plot(df['Close'].to_list(), marker="o", markersize=1, color="black", linewidth=1, markeredgecolor="black", label = 'Close Price' )
plt.plot(sentiment_score, marker="o", markersize=1, color="purple", linewidth=1, markeredgecolor="purple", label = 'Sentiment Score' )
plt.plot(change, marker="o", markersize=1, color="indigo", linewidth=1, markeredgecolor="indigo", label = 'Delta' )
plt.title('Tweet Sentiment and Stock Price Movement')
plt.ylabel('Stock Price')
plt.xlabel('Time Series [2014-2015]')
plt.legend()
plt.show()
```

```python
plt.rcParams["figure.figsize"] = [10.00, 5.00]
plt.rcParams["figure.autolayout"] = True

# plt.xlim(xlim[0], xlim[1])
# plt.ylim(ylim[0], ylim[1])
plt.grid()

plt.plot(df['Close'].to_list(), marker="o", markersize=1, color="black", linewidth=1, markeredgecolor="black", label = 'Close Price' )
plt.plot(change, marker="o", markersize=1, color="purple", linewidth=1, markeredgecolor="purple", label = 'Delta' )
plt.plot(total_tweets, marker="o", markersize=1, color="indigo", linewidth=1, markeredgecolor="indigo", label = 'Total Tweets' )
plt.title('Tweets Volume and Stock Price Movement')
plt.ylabel('Stock Price')
plt.xlabel('Time Series [2014-2015]')
plt.legend()
plt.show()
```
